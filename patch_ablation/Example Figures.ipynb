```python
import os
import sys
import torch
sys.path.append('../src')
import ds_utils
import torchvision
dataset = ds_utils.ImageNetMegaDS(factor=1000, do_mask=True)
```

```python
def display_and_save(img, filename):
    pil = torchvision.transforms.ToPILImage()(img)
    pil.save(filename)
    display(pil)
```

```python
img = dataset[8]['image'].clone()
display_and_save(img, 'lady_dog1.pdf')
mask = dataset[8]['mask']
mask = torch.ones_like(img)
mask[:, :, 10:80] = 0
img[mask == 0] = 0 
display_and_save(img, 'lady_dog2.pdf')
img = dataset[8]['image'].clone()
mask = dataset[8]['mask']
mask = torch.ones_like(img)
mask[:, 40:90, 80:120] = 0
img[mask == 0] = 0 
display_and_save(img, 'lady_dog3.pdf')
```
